```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mtcars = pd.read_csv('mtcars.csv')
mtcars
```

```python
from sklearn import linear_model

regression_model = linear_model.LinearRegression()

regression_model.fit(X = pd.DataFrame(mtcars['wt']), y = mtcars['mpg'])
regression_model.predict(X = pd.DataFrame(mtcars['wt']))
regression_model.coef_
```

```python
# Prueba de R^2. Mejor si está más cerca de 1
regression_model.score(X = pd.DataFrame(mtcars['wt']), y = mtcars['mpg'])
```

```python
# Se puede calcular R2 manualmente
R2 = 1- np.sum(mtcars['mpg']-regression_model.predict(X = mtcars['wt']))/np.sum((mtcars['mpg']-np.mean(mtcars['mpg']))**2)
```
