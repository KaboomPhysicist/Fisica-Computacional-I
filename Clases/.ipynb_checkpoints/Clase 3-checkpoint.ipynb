## Método de Euler
 La ecuación que se usa de forma iterativa es $y_{m+1} = y_m + hf(t_m,y_m)$.

#### Problema 1. Ecuaciones cinemáticas

Se parte de las ecuaciones diferenciales:

\begin{equation}
\frac{dx}{dt}=V_{1x} \\
\frac{dy}{dt}=V_{1y} - gt
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81

fx = lambda Vx: Vx
fy = lambda Vy, t0: Vy - g*t0

def Euler(inicio, delta):
    vel_ini = inicio[0]
    ang = np.radians(inicio[1])
    
    # Sirve para determinar el tiempo que tarda en volver al suelo
    ts = vel_ini*np.sin(ang)/g
    
    T = np.arange(0,2*ts,delta)
    X = np.zeros(len(T))
    Y = np.zeros(len(T))
    
    X[0] = 0
    Y[0] = 0
    
    # Aplicar Euler
    for i in range(len(T)-1):
        X[i+1] = X[i] + delta*vel_ini*np.cos(ang)
        Y[i+1] = Y[i] + delta*fy(vel_ini*np.sin(ang), T[i])
    
    # Solución analítica del problema
    XT = vel_ini*np.cos(ang)*T
    YT = vel_ini*np.sin(ang)*T-0.5*g*T**2
    
    plt.figure(figsize=(16,9))
    plt.xlabel("x [m]")
    plt.ylabel("y [m]")
    plt.plot(X, Y, label = "Solución numérica (Euler)")
    plt.plot(XT, YT, label = "Solución analítica")
    plt.legend()
    plt.grid()
    
    return X,Y
```

```python
Euler((50,53.1), 1)
```

## Método de Euler mejorado

Las ecuaciones para aplicar el método son:

\begin{equation}
y_{n+1} = y_n + \frac{h}{2} (f(x_n,y_n)+f(x_{n+1},y^*_{n+1})) \\
y^*_{n+1} = y_n + hf(x_n,y_n)
\end{equation}

```python
g = 9.81

fx = lambda Vx: Vx
fy = lambda Vy, t0: Vy - g*t0

def Euler_improv(inicio, delta):
    vel_ini = inicio[0]
    ang = np.radians(inicio[1])
    
    # Sirve para determinar el tiempo que tarda en volver al suelo
    ts = vel_ini*np.sin(ang)/g
    
    T = np.arange(0,2*ts,delta)
    X = np.zeros(len(T))
    Y = np.zeros(len(T))
    
    X[0] = 0
    Y[0] = 0
    
    # Aplicar Euler
    for i in range(len(T)-1):
        X[i+1] = X[i] + delta*vel_ini*np.cos(ang)
        yaux = Y[i+1] + delta*fy(vel_ini*np.sin(ang), T[i+1])
        Y[i+1] = Y[i] + 0.5*delta*(fy(vel_ini*np.sin(ang), T[i])+yaux)
    
    # Solución analítica del problema
    XT = vel_ini*np.cos(ang)*T
    YT = vel_ini*np.sin(ang)*T-0.5*g*T**2
    
    plt.figure(figsize=(16,9))
    plt.xlabel("x [m]")
    plt.ylabel("y [m]")
    plt.plot(X, Y, label = "Solución numérica (Euler)")
    plt.plot(XT, YT, label = "Solución analítica")
    plt.legend()
    plt.grid()
    
    return X,Y
```

```python
Euler_improv((50,53.1), 0.001)
```
