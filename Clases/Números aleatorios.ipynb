```python
def random_manual(N, seed):
    a = 57
    c = 1
    M = 256
    r0 = seed
    
    ranlist = []
    for i in range(N):
        ranlist.append(r0)
        r0 = (a*r0+c)%M
        #print(r0)
        
    return ranlist
    
```

```python
random_manual(15, 3)
```

### Distribución exponencial

```python
import numpy as np

l = 3

def dist_exp(N):
    rand_dist = np.zeros(N)
    for i in range(N):
        rand_dist[i] = -l*np.log(1 - random.random())
    
    return rand_dist
        
```

```python
plt.hist(dist_exp(100000), bins=10, density = True)
plt.show()
```

### Distribución gaussiana

```python
# Box-Muller
def dist_gauss(N):
    r1 = np.random.rand(N)
    r2 = np.random.rand(N)
    
    R = np.sqrt(-2*np.log(r1))*np.cos(2*np.pi*r2)
    
    return R
    
```

```python
plt.hist(dist_gauss(100000), bins=100, density = True)
plt.grid()
```

```python
# Usando numpy
plt.hist(np.random.normal(size = 100000), bins = 100, density = True)
plt.grid()
```

```python
# Teorema del límite central

def gauss_central(N):
    rnn = np.zeros(N)
    for i in range(N):
        rnn[i] = np.sum(np.random.random(1000))
    # Estandariza el resultado
    return (rnn-rnn.mean())/rnn.std()
```

```python
plt.hist(gauss_central(100000), bins = 100, density = True)
plt.grid()
```
