```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

import seaborn as sns

data = pd.read_csv('Birthweight_reduced_kg_R.csv')
data
```

#### 2. Contabilice las variables categóricas y grafíquelas.

Hay 3 variables categóricas: smoker, lowbwt y mage35.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (16,9))
data["smoker"].value_counts().plot(kind = 'pie', ax = axes[0])
data["lowbwt"].value_counts().plot(kind = 'pie', ax = axes[1])
data["mage35"].value_counts().plot(kind = 'pie', ax = axes[2])
```

#### 3. ¿El peso del bebé se distribuye normalmente? ¿el número de cigarrillos fumados por día también se distribuye normalmete?

```python
from scipy.stats import normaltest

mean_bwt = data['Birthweight'].mean()
std_bwt = data['Birthweight'].std()
plt.hist(data['Birthweight'], density = True)
plt.plot(np.linspace(2,4.5,100), norm.pdf(np.linspace(2,4.5,100), loc = mean_bwt, scale = std_bwt))
#print(mean_bwt, data['Birthweight'].median())


# normaltest prueba un ajuste y determina el p-value por una distribución chi cuadrado.
# Si el p-value es menor a la significancia alpha, la hipótesis nula (la distribución es normal)
# se puede rechazar.
normaltest(data['Birthweight'])
```

En este caso, el pvalue obtenido es 0.949, si tomamos una significancia de 0.05, entonces la hipótesis nula no puede ser rechazada, i.e, los datos siguen una distribución normal.

```python
mean_cig = data['mnocig'].mean()
std_cig = data['mnocig'].std()
plt.hist(data['mnocig'], density = True)
plt.plot(np.linspace(0, 50,100), norm.pdf(np.linspace(2,4.5,100), loc = mean_cig, scale = std_cig))

normaltest(data['mnocig'])
```

Los datos no siguen una distribución normal.
